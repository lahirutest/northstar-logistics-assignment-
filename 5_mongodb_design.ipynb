AGGREGATION PIPELINES

MongoDB aggregation pipelines support:
- operational intelligence
- real-time analytics
- scalable operational reporting

```python
# ============================================================
# STEP 13 — AGGREGATION:
# FAILED DELIVERIES BY HUB
# ============================================================

pipeline = [
    {
        "$match":
        {
            "delivery_status": "Failed"
        }
    },
    {
        "$group":
        {
            "_id": "$hub_id",
            "failed_deliveries":
            {
                "$sum": 1
            }
        }
    },
    {
        "$sort":
        {
            "failed_deliveries": -1
        }
    }
]

results = deliveries_collection.aggregate(pipeline)

for r in results:
    print(r)
```

INTERPRETATION:

This aggregation pipeline identifies operational hubs
with the highest failed delivery counts.

MongoDB aggregation supports scalable operational analytics
across large event-driven datasets.

```python
# ============================================================
# STEP 14 — AGGREGATION:
# HIGH LATENCY APP EVENTS
# ============================================================

pipeline2 = [
    {
        "$group":
        {
            "_id": "$event_type",
            "avg_latency":
            {
                "$avg": "$api_latency_ms"
            }
        }
    },
    {
        "$sort":
        {
            "avg_latency": -1
        }
    }
]

results2 = app_events_collection.aggregate(pipeline2)

for r in results2:
    print(r)
```

INTERPRETATION:

This aggregation evaluates digital platform performance
across customer interaction events.

High API latency may negatively affect
customer experience and operational responsiveness.

MongoDB embedded documents support:
- complete operational histories
- nested complaint tracking
- customer interaction timelines

This reduces expensive relational joins
and improves operational query performance.
